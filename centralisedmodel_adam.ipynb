```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
# --- ADDED: Import the LR scheduler ---
from torch.optim.lr_scheduler import ReduceLROnPlateau
# --- END OF ADDITION ---

# -------------------------------------------------------------------
# CONFIGURATION
# -------------------------------------------------------------------

# --- Path to your main manifest file ---
MANIFEST_PATH = "/home/DSE423/Documents/FL Project/manifest.csv"

# --- Centralized split files ---
TRAIN_CSV_PATH = "central_train.csv"
VAL_CSV_PATH = "central_val.csv"
TEST_CSV_PATH = "central_test.csv"

# --- Model & Training Parameters ---
BATCH_SIZE = 32
NUM_EPOCHS = 50
MOMENTUM = 0.9 # Kept for SGD, but AdamW (below) doesn't use it
BEST_MODEL_PATH = "centralized_best_model.pth"
HISTORY_PLOT_PATH = "training_history.png"
LEARNING_RATE = 0.0001 

# --- Define class names (MUST match your notebook) ---
CLASS_NAMES = ['Non Demented', 'Very mild Dementia', 'Mild Dementia', 'Moderate Dementia']

# -------------------------------------------------------------------
# STEP 1: DATA SPLITTING
# -------------------------------------------------------------------
def create_centralized_splits(manifest_path):
    print("--- STEP 1: Creating Data Splits ---")
    # Check if files already exist to avoid re-splitting
    if os.path.exists(TRAIN_CSV_PATH) and os.path.exists(VAL_CSV_PATH) and os.path.exists(TEST_CSV_PATH):
        print("CSV split files already exist. Loading them.")
        train_df = pd.read_csv(TRAIN_CSV_PATH)
        val_df = pd.read_csv(VAL_CSV_PATH)
        test_df = pd.read_csv(TEST_CSV_PATH)
        print("âœ… Data splits loaded.")
    else:
        print(f"Loading master manifest from {manifest_path}...")
        try:
            df = pd.read_csv(manifest_path)
        except FileNotFoundError:
            print(f"âŒ ERROR: File not found at {manifest_path}")
            return False
        
        print(f"Loaded {len(df)} total records.")
        if 'label' not in df.columns:
            print("âŒ ERROR: 'label' column not found in manifest.csv")
            return False
        
        print("Splitting data into stratified train (70%), val (15%), test (15%)...")
        y = df['label']
        train_df, temp_df = train_test_split(df, test_size=0.3, stratify=y, random_state=42)
        val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label'], random_state=42)
        
        train_df.to_csv(TRAIN_CSV_PATH, index=False)
        val_df.to_csv(VAL_CSV_PATH, index=False)
        test_df.to_csv(TEST_CSV_PATH, index=False)
        print("âœ… Data splitting successful!")

    print(f"  Training set:   {len(train_df)} images")
    print(f"  Validation set: {len(val_df)} images")
    print(f"  Test set:       {len(test_df)} images")
    return True

# -------------------------------------------------------------------
# STEP 2: DEFINE THE DATASET CLASS
# -------------------------------------------------------------------
class MRIDataset(torch.utils.data.Dataset):
    def __init__(self, csv_path, transform=None):
        self.data = pd.read_csv(csv_path)
        self.transform = transform
        self.image_col = 'image_path' 
        self.label_col = 'label'

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data.loc[idx, self.image_col]
        label_str = self.data.loc[idx, self.label_col]
        
        try:
            image = Image.open(img_path).convert("RGB")
        except FileNotFoundError:
            warnings.warn(f"Warning: Image not found at {img_path}. Returning None.")
            return None 

        if self.transform:
            image = self.transform(image)
            
        label_map = {
            "Non Demented": 0,
            "Very mild Dementia": 1,
            "Mild Dementia": 2,
            "Moderate Dementia": 3,
        }
        label = label_map[label_str]
        
        return image, label

# -------------------------------------------------------------------
# STEP 3: DEFINE THE MODEL
# -------------------------------------------------------------------
class MRI_CNN(nn.Module):
    def __init__(self, num_classes=4):
        super(MRI_CNN, self).__init__()
        self.block1 = nn.Sequential(
            nn.Conv2d(3, 32, 3, 1, 1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.block3 = nn.Sequential(
            nn.Conv2d(64, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.MaxPool2d(2),
        )
        self.block4 = nn.Sequential(
            nn.Conv2d(128, 256, 3, 1, 1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Dropout(0.4),
        )
        self.global_avg_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.classifier = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes),
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        x = self.global_avg_pool(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

# -------------------------------------------------------------------
# STEP 4: DEFINE TRANSFORMS
# -------------------------------------------------------------------
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3),
])

# -------------------------------------------------------------------
# STEP 5: TRAINING & VALIDATION FUNCTIONS
# -------------------------------------------------------------------
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct_preds = 0
    total_preds = 0
    
    for (images, labels) in tqdm(dataloader, desc="Training", leave=False):
        if images is None or labels is None or images.numel() == 0:
            continue
            
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total_preds += labels.size(0)
        correct_preds += (predicted == labels).sum().item()

    epoch_loss = running_loss / total_preds if total_preds > 0 else 0
    epoch_acc = correct_preds / total_preds if total_preds > 0 else 0
    return epoch_loss, epoch_acc

def validate(model, dataloader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct_preds = 0
    total_preds = 0
    
    with torch.no_grad():
        for (images, labels) in tqdm(dataloader, desc="Validation", leave=False):
            if images is None or labels is None or images.numel() == 0:
                continue

            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total_preds += labels.size(0)
            correct_preds += (predicted == labels).sum().item()

    epoch_loss = running_loss / total_preds if total_preds > 0 else 0
    epoch_acc = correct_preds / total_preds if total_preds > 0 else 0
    return epoch_loss, epoch_acc

# -------------------------------------------------------------------
# STEP 6: HELPER FUNCTIONS (Collate, Plotting)
# -------------------------------------------------------------------
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    if not batch:
        return None, None
    return torch.utils.data.dataloader.default_collate(batch)

def plot_confusion_matrix_heatmap(cm, class_names):
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm, 
        annot=True, 
        fmt='d', 
        cmap='Blues', 
        xticklabels=class_names, 
        yticklabels=class_names
    )
    plt.title('Centralized Model Confusion Matrix (Test Set)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig('centralized_confusion_matrix.png')
    print("Confusion matrix saved to 'centralized_confusion_matrix.png'")

def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))
    fig.suptitle('Centralized Model Training History', fontsize=16, y=1.02)
    
    # Plot Loss
    ax1.plot(history['train_loss'], label='Train Loss')
    ax1.plot(history['val_loss'], label='Validation Loss')
    ax1.set_title('Model Loss per Epoch')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    # Plot Accuracy
    ax2.plot(history['train_acc'], label='Train Accuracy')
    ax2.plot(history['val_acc'], label='Validation Accuracy')
    ax2.set_title('Model Accuracy per Epoch')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.savefig(HISTORY_PLOT_PATH)
    print(f"Training history plot saved to '{HISTORY_PLOT_PATH}'")

# -------------------------------------------------------------------
# STEP 7: MAIN TRAINING SCRIPT
# -------------------------------------------------------------------
def run_training_pipeline():
    print("\n--- STEP 2: Setting up Training ---")
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {DEVICE}")

    print("Loading datasets from split CSVs...")
    train_dataset = MRIDataset(csv_path=TRAIN_CSV_PATH, transform=transform)
    val_dataset = MRIDataset(csv_path=VAL_CSV_PATH, transform=transform)
    
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, collate_fn=collate_fn)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, collate_fn=collate_fn)
    
    print("Initializing model...")
    model = MRI_CNN().to(DEVICE)

    # --- ADDED: Class Weight Calculation (Change 1) ---
    print("Calculating class weights to handle imbalance...")
    # Counts based on 70% of totals from manifest (Non: 67222, VMD: 13725, MD: 5002, MOD: 488)
    count_non = 47055
    count_vmid = 9608
    count_mid = 3501
    count_mod = 342
    total = count_non + count_vmid + count_mid + count_mod

    # Formula: total / (num_classes * count)
    weights = [
        total / (4 * count_non),
        total / (4 * count_vmid),
        total / (4 * count_mid),
        total / (4 * count_mod) 
    ]
    class_weights = torch.tensor(weights).float().to(DEVICE)
    print(f"Using class weights (Non, VMD, MD, MOD): {np.round(weights, 2)}")
    # --- END OF ADDITION ---
    
    # --- MODIFIED LINE (Change 1) ---
    criterion = nn.CrossEntropyLoss(weight=class_weights)
    # --- END OF MODIFICATION ---
    
    # --- MODIFIED: Switched to AdamW optimizer (Change 3) ---
  # --- MODIFIED: Switched to AdamW optimizer for stability ---
    optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE)
    print(f"Using optimizer: AdamW (lr={LEARNING_RATE})")
# --- END OF MODIFICATION ---
    # --- END OF MODIFICATION ---
    
    # --- ADDED: Learning Rate Scheduler (Change 2, with verbose=True removed) ---
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.1, patience=3)
    print("Using ReduceLROnPlateau LR scheduler.")
    # --- END OF ADDITION ---
    
    best_val_acc = 0.0
    
    # History dictionary to store metrics
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }

    print(f"\n--- STEP 3: Starting Centralized Training ({NUM_EPOCHS} Epochs) ---")
    for epoch in range(NUM_EPOCHS):
        print(f"\n--- Epoch {epoch+1}/{NUM_EPOCHS} ---")
        
        train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)
        val_loss, val_acc = validate(model, val_loader, criterion, DEVICE)
        
        # --- ADDED: Scheduler Step (Change 2) ---
        scheduler.step(val_loss)
        # --- END OF ADDITION ---
        
        print(f"Epoch {epoch+1} Summary:")
        print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc*100:.2f}%")
        print(f"  Val Loss: {val_loss:.4f}   | Val Acc: {val_acc*100:.2f}%")
        
        # Store history
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        # Save the best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), BEST_MODEL_PATH)
            print(f"  ðŸŽ‰ New best model saved with Val Acc: {best_val_acc*100:.2f}%")

    print("--- Training Finished ---")
    
    # Plot the history
    plot_training_history(history)
    
    # -------------------------------------------------------------------
    # STEP 8: FINAL TESTING
    # -------------------------------------------------------------------
    print(f"\n--- STEP 4: Final Testing (on held-out test set) ---")
    print(f"Loading best model from {BEST_MODEL_PATH}...")
    
    # Load test set
    test_dataset = MRIDataset(csv_path=TEST_CSV_PATH, transform=transform)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, collate_fn=collate_fn)

    # Load best model weights
    # Re-initialize model first, then load state
    model = MRI_CNN().to(DEVICE) 
    model.load_state_dict(torch.load(BEST_MODEL_PATH))
    model.eval()
    
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for (images, labels) in tqdm(test_loader, desc="Testing"):
            if images is None or labels is None or images.numel() == 0:
                continue 
            
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())
    
    # Calculate final metrics
    test_acc = accuracy_score(all_labels, all_preds)
    print(f"\n========================================================")
    print(f"ðŸ“Š FINAL TEST ACCURACY: {test_acc*100:.2f}%")
    print(f"========================================================")
    
    # Generate and plot confusion matrix
    cm = confusion_matrix(all_labels, all_preds, labels=[0, 1, 2, 3])
    plot_confusion_matrix_heatmap(cm, class_names=CLASS_NAMES)
    
    print("\n--- Centralized Baseline Run Complete ---")


# -------------------------------------------------------------------
# SCRIPT EXECUTION
# -------------------------------------------------------------------
if __name__ == "__main__":
    # Suppress UserWarnings from PIL
    warnings.filterwarnings("ignore", category=UserWarning, module="PIL")
    
    # First, create the split files from the main manifest
    if create_centralized_splits(MANIFEST_PATH):
        # If splitting was successful, run the training
        run_training_pipeline()
    else:
        print("Halting execution due to error in data splitting.")
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, confusion_matrix
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings

# -------------------------------------------------------------------
# CONFIGURATION (Redefining for this new cell)
# -------------------------------------------------------------------
BATCH_SIZE = 32
BEST_MODEL_PATH = "centralized_best_model.pth"
TEST_CSV_PATH = "central_test.csv"
CLASS_NAMES = ['Non Demented', 'Very mild Dementia', 'Mild Dementia', 'Moderate Dementia']
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -------------------------------------------------------------------
# ADNET MODEL (Replaces old CNN)
# -------------------------------------------------------------------
class ADNet(nn.Module):
    def __init__(self, num_classes=4):
        super(ADNet, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc1 = nn.Linear(128 * 16 * 16, 256)
        self.fc2 = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)
        x = nn.ReLU()(self.fc1(x))
        x = self.fc2(x)
        return x

# -------------------------------------------------------------------
# DATASET/TRANSFORM DEFINITIONS
# -------------------------------------------------------------------
class MRIDataset(torch.utils.data.Dataset):
    def __init__(self, csv_path, transform=None):
        self.data = pd.read_csv(csv_path)
        self.transform = transform
        self.image_col = 'image_path'
        self.label_col = 'label'

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data.loc[idx, self.image_col]
        label_str = self.data.loc[idx, self.label_col]

        try:
            image = Image.open(img_path).convert("RGB")
        except FileNotFoundError:
            warnings.warn(f"Warning: Image not found at {img_path}. Returning None.")
            return None

        if self.transform:
            image = self.transform(image)

        label_map = {
            "Non Demented": 0,
            "Very mild Dementia": 1,
            "Mild Dementia": 2,
            "Moderate Dementia": 3,
        }
        label = label_map[label_str]

        return image, label

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3),
])

# -------------------------------------------------------------------
# HELPER FUNCTIONS
# -------------------------------------------------------------------
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    if not batch:
        return None, None
    return torch.utils.data.dataloader.default_collate(batch)

def plot_confusion_matrix_heatmap(cm, class_names):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title('Centralized Model Confusion Matrix (Test Set)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig('centralized_confusion_matrix.png')
    print("Confusion matrix saved to 'centralized_confusion_matrix.png'")

# -------------------------------------------------------------------
# CENTRALIZED TRAINING SCRIPT
# -------------------------------------------------------------------

def train_centralized(train_csv, val_csv, num_epochs=20, lr=0.0001):
    print("
--- Starting Centralized Training ---")

    train_dataset = MRIDataset(csv_path=train_csv, transform=transform)
    val_dataset = MRIDataset(csv_path=val_csv, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, collate_fn=collate_fn)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, collate_fn=collate_fn)

    model = ADNet().to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    best_val_acc = 0.0

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        correct = 0
        total = 0

        for images, labels in train_loader:
            if images is None: continue
            images, labels = images.to(DEVICE), labels.to(DEVICE)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_acc = correct / total

        # Validation
        model.eval()
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                if images is None: continue
                images, labels = images.to(DEVICE), labels.to(DEVICE)
                outputs = model(images)
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        val_acc = val_correct / val_total

        print(f"Epoch [{epoch+1}/{num_epochs}] | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), BEST_MODEL_PATH)
            print(f"âœ¨ Saved Best Model (Val Acc: {best_val_acc:.4f})")

    print("
--- Training Complete ---")
    print(f"Best Validation Accuracy: {best_val_acc:.4f}")

# -------------------------------------------------------------------
# FINAL TESTING SCRIPT
# -------------------------------------------------------------------
print(f"
--- Running Final Test ---")
print(f"Loading best model from {BEST_MODEL_PATH}...")
print(f"Using device: {DEVICE}")

warnings.filterwarnings("ignore", category=UserWarning, module="PIL")

try:
    test_dataset = MRIDataset(csv_path=TEST_CSV_PATH, transform=transform)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, collate_fn=collate_fn)

    model = ADNet().to(DEVICE)
    model.load_state_dict(torch.load(BEST_MODEL_PATH, map_location=DEVICE))
    model.eval()

    all_labels = []
    all_preds = []

    with torch.no_grad():
        for (images, labels) in tqdm(test_loader, desc="Testing"):
            if images is None or labels is None or images.numel() == 0:
                continue

            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())

    test_acc = accuracy_score(all_labels, all_preds)
    print(f"
========================================================")
    print(f"ðŸ“Š FINAL TEST ACCURACY: {test_acc*100:.2f}%")
    print(f"========================================================")

    cm = confusion_matrix(all_labels, all_preds, labels=[0, 1, 2, 3])
    plot_confusion_matrix_heatmap(cm, class_names=CLASS_NAMES)

    print("
--- Centralized Baseline Test Complete ---")

except FileNotFoundError:
    print(f"âŒ ERROR: Could not find '{BEST_MODEL_PATH}' or '{TEST_CSV_PATH}'.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")import torch
import torch.nn as nn
import torch.nn.functional as F

class ADNet(nn.Module):
    def __init__(self, num_classes=2):
        super(ADNet, self).__init__()

        # Block 1
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        
        # Block 2
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(64)

        # Block 3
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        # Fully connected layers
        self.fc1 = nn.Linear(128 * 16 * 16, 256)  # assuming input is 1x128x128
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(256, 64)
        self.fc3 = nn.Linear(64, num_classes)

    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.max_pool2d(x, 2)

        x = F.relu(self.bn2(self.conv2(x)))
        x = F.max_pool2d(x, 2)

        x = F.relu(self.bn3(self.conv3(x)))
        x = F.max_pool2d(x, 2)

        x = x.view(x.size(0), -1)

        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)

        return x

# Example usage
# model = ADNet(num_classes=2)
# print(model)
```
